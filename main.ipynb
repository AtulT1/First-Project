```python
#Importing Libs
import pandas as pd
from numpy import unique,loadtxt
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as snc
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
```

```python
# Dropping Data if NAN Value is greater than 80% and check duplication
df = pd.read_csv("train.csv") #Reading Data
Deleted_Col=set()
for col in df.columns:
    x=df[col].isnull().sum()/df.shape[0]*100 #percent of null values
    
    if x!=0:
        if x>80:
            Deleted_Col.add(f"{col}")
            df.drop(col,axis=1,inplace=True)
        else:
            plt.scatter(col,x)
            plt.xticks(rotation=90)
            plt.grid(visible=True)
            plt.axhline(y=80) #Threesold
print(f"Dropped Column due to empty vals which are more than 80%:{Deleted_Col}")
duplicate=df[df.duplicated()]
print(f"Number of duplicate rows : {duplicate.shape[0]}")

# plt.show()
```

```python
#Mappping Catgeorycle data
for col in  df.columns:

    if df[col].dtypes=='object':
        mapped_data=df[col].unique()
        # print(type(mapped_data))
        # print(df[col].value_counts())
        x={}
        count=0
        for i in range(len(mapped_data)):
            x[f"{mapped_data[i]}"]=count
        #     mapped_data[i]=count
            count+=1
        df[col]=df[col].map(x)
        # print(df[col])
        

        # print(mapped_data)
    # print(mapped_data)
    # print(df)
    
            
```

```python
#Spliting Data for testing
# df.info()
#df.drop('Id',inplace=True,axis=1)

a=df.drop('SalePrice',axis=1)
b=df['SalePrice']
a_train, a_test, b_train, b_test = train_test_split(a,b,test_size=0.20,random_state=35)
print(len(a_train),len(a_test),len(b_train),len(b_test))
```

```python
#model Creation
dict1=DecisionTreeRegressor(random_state=45)
dict1.fit(a_train,b_train)
```

```python
#Testing Feature Importance
feature_imp=pd.Series(dict1.feature_importances_,index=a.columns)
feature_imp.nlargest(15).plot(kind="barh")
plt.show()
```
